## Determining Most Popular Topics

```python
with open('mastodon_instance_data.json') as json_file:
    myData = json.load(json_file)
```

```python
topics = []
for i in myData:
    if 'info' in i.keys():
        if i['info'] and 'topic' in i['info']:
            topics.append(i['info']['topic'])
```

```python
stop_words = set(stopwords.words('english'))
```

```python
valid_topics = [i for i in topics if i] 
```

```python
keywords = []
for i in valid_topics:
    words = i.split(", ")
    for word in words:
        keywords.append(word.lower())
```

```python
single_keywords = []
for i in keywords:
    words = i.split()
    for word in words:
        if word not in stop_words: single_keywords.append(word)
```

```python
keywordsCounter = Counter(keywords)
keywordsCounter.most_common()

singleCounter = Counter(single_keywords)
singleCounter.most_common()
```

```python
topic_count = singleCounter.most_common()
```

```python
with open('mastodon_topics.csv', 'w') as out:
    csv_out = csv.writer(out)
    csv_out.writerow(['topic', 'count'])
    for topic in topic_count:
        csv_out.writerow(topic)
```
